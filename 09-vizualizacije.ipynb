## Matplotlib

- [Spletna stran](https://matplotlib.org/)
- [Galerija primerov](https://matplotlib.org/stable/gallery/index.html)
- [Dokumentacija (API Reference)](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.html)

Python knjižnica za izris grafov, histogramov, stolpičnih in krožnih diagramov, razsevnih diagramov, ...

```python
import matplotlib.pyplot as plt
import numpy as np
```

Knjižnico tipično uvozimo pod imenom `plt`. Najosnovnejšo uporabo bomo demostrirali z izrisom kvadratne funkcije. S funkcijo `plot` bomo izrisali **graf funkcije**, ki povezuje podane točke, ki so definirane s seznamoma x in y-koordinat. Izrisu lahko nastavimo tudi oznake osi in mu dodamo naslov.

```python
xs = range(-3, 5)
ys = [x**2 for x in xs]
plt.plot(xs, ys)
plt.title("Kvadratna funkcija")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

Matplotlib se dobro ujame tudi z `numpy` knjižnico. Zgenerirajmo množico števil z normalno porazdelitvijo in prikažimo vzorec s **stolpičnim diagramom** (*bar chart*). Števila razdelimo v 10  enako širokih skupin in izrišimo histogram.

```python
np.random.seed(12345)
xs = np.random.normal(size=1000)
plt.bar(range(1000), xs)
plt.show()
plt.hist(xs, bins=10)
plt.show()
```

Pripravimo si podatke o starosti, plači in spolu izmišljenih oseb.

```python
n = 20
starost = np.random.randint(20, 70, size=n)
placa = np.random.randint(500, 3000, size=n)
spoli = ["moški", "ženska"]
spol = np.random.choice(spoli, n)
```

Podatke bomo prikazali na **razsevnem diagramu** (*scatter plot*). Na x-osi naj bo starost, na y-osi plača, spol pa bomo prikazali z barvo. Tudi za seznam barv bomo uporabili numpy tabelo, kar nam bo prišlo prav kasneje.

```python
color = np.array(["red" if s=="ženska" else "blue" for s in spol])
plt.scatter(starost, placa, c=color)
plt.show()
```

Vizualizacijo opremimo z mrežo črtkanih črt za boljšo preglednost. Dodajmo naslove osem in diagramu. Narišimo še črto, ki označuje povprečno plačo in dodajmo anotacijo, kaj črta predstavlja.

```python
plt.title("Razsevni diagram zaslužka glede na starost")
plt.ylabel("Plača")
plt.xlabel("Starost")
plt.grid(linestyle="--")

plt.scatter(starost, placa, c=color)
avg = np.mean(placa)
plt.plot([20, 70], [avg, avg], linestyle=":", c="black")
plt.annotate("povprečje", (70-10, avg+100))

plt.show()
```

Nazadnje dodajmo še legendo barv. Za to bomo morali narisati več diagramov, vsakega s svojo oznako. Z vsak spol bomo ustvarili masko, ki nam označuje, kateri primeri pripadajo temu spolu, da izrišemo so njih. Na koncu pa bomo lahko dorisali še legendo oznak. Izrisan diagram bomo tudi shranili kot sliko za vključitev v kakšno poročilo ali zgolj kasnejši ogled. Končna rešitev izgleda tako.

```python
plt.title("Razsevni diagram zaslužka glede na starost")
plt.ylabel("Plača")
plt.xlabel("Starost")
plt.grid(linestyle="--")

for s in spoli:
    mask = (spol == s)
    plt.scatter(starost[mask], placa[mask], c=color[mask], label=s)

avg = np.mean(placa)
plt.plot([20, 70], [avg, avg], linestyle=":", c="black")
plt.annotate("povprečje", (70-10, avg+100))

plt.legend(loc="upper right")
plt.savefig("placa.png")
plt.show()
```

### Slike

Knjižnica matplotlib omogoča tudi prikazovanje slik, ki niso nič drugega kot tabela vrednosti posameznih pikslov na sliki. Za primer bomo naložili slike [ročno narisanih števk](https://archive.ics.uci.edu/dataset/80/optical+recognition+of+handwritten+digits) s pomočjo knjižnice [scikit-learn](https://scikit-learn.org/stable/), ki omogoča enostaven dostop do te podatkovne zbirke s funkcijo [`load_digits`](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_digits.html). Piksli na sliki so po vrsticah zloženi v tabelo dolžine 64, ki jo bomo za prikaz preoblikovali v 2D tabelo dimenzij 8x8. Za prikaz slike uporabimo funkcijo `imshow`. Kateri paleti barv ustrezajo vrednosti v tabeli pa določa argument `cmap` (color map).

```python
import sklearn.datasets

digits = sklearn.datasets.load_digits()['data']
digit = digits[1234]
print(digit.shape)
digit = digit.reshape(8,8)
print(digit)
plt.figure(figsize=(2,2))
plt.imshow(digit, cmap="gray")
plt.show()
```

Prikažimo sedaj več števk z uporabo `subplots` za izris več grafov na istem prikazu. Funkciji podamo število vrstic in stolpcev, vrne pa nam dva objekta. Prvi predstavlja celoten prikaz, drugi pa tabelo območij (axes). Z vsakim območjem lahko delamo podobno, kot smo prej s celim grafom - odstranimo mu oznake na oseh in na njem prikažemo sliko števke. Ustvarjen prikaz še shranimo v png sliko.

```python
h, w = 2, 5
n = h*w
fig, axes = plt.subplots(h, w, figsize=(7,3))
fig.suptitle("Digits")  # super title
for i in range(h):
    for j in range(w):
        ax = axes[i,j]
        ax.set_xticks([])
        ax.set_yticks([])
        ax.imshow(digits[i*w+j].reshape(8,8), cmap="gray")
plt.savefig('digits.png')
```

Sestavimo še kakšno barvno sliko. To bo tabela dimenzije $H \times W \times 3$. Definirati moramo tudi tip vrednosti. Lahko bi hranili zvezne vrednosti med 0.0 in 1.0 ali pa celoštevilske med 0 in 255. Slednje je bolj običajen primer, ki mu ustreza celoštevilski tip `uint8`. Izbrali si bomo lokacije izvorov rdeče, zelene in modre barve. Piksli na sliki pa bodo dobili na vsakem kanalu (rdeča, zelena, modra) intenziteto, ki je sorazmerna z oddaljenostjo od izvora tiste barve. Sliko bomo shranili in prikazali.

```python
h, w = 200, 300
source = [(0,300), (50,40), (150,180)]  # (vrstica, stolpec) izvora rdece, zelene in modre
slika = np.zeros((h,w,3), dtype=np.uint8)
for y in range(h):
    for x in range(w):
        for c in range(3):
            d = ((y-source[c][0])**2 + (x-source[c][1])**2)**0.5
            slika[y,x,c] = max(int(255-d),0)
plt.imsave("mavrica.png", slika)

plt.figure(figsize=(4,6))
plt.axis("off")
plt.imshow(slika)
plt.show()
```
